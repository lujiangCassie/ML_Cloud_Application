```python
import pandas as pd
import numpy as np
```

```python
datafile= './data/census.csv'
```

```python
df = pd.read_csv(datafile)
df.head(10)
```

```python
df.info()
```

```python
df.isna().sum(axis=0)
# from the results, there is not missing value cells
```

```python
df.describe()  # statistics for numeric columns
```

```python
df.columns = [c.strip() for c in df.columns]
df.info()
```

```python
categorical_features_indices = np.where(df.dtypes == object)[0]
cat_features = df.columns[categorical_features_indices].tolist()
print(cat_features)
```

```python
for c in cat_features:
    df[c] = df[c].str.strip()
```

```python
for c in cat_features:
    print(df[c].value_counts())
```

```python
df_clean = df.replace('?', None).dropna()
```

```python
for c in cat_features:
    print(df_clean[c].value_counts())
```

```python
df_clean.to_csv("./data/census.csv", index=False)
```
